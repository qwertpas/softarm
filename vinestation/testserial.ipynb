```python
import serial
import time
```

```python
port = '/dev/tty.usbmodem1'
serial_port = serial.Serial(port, 921600, timeout=0.1)
```

```python
msg = f"P10:1\n"
serial_port.write(msg.encode('utf-8'))
time.sleep(0.5)
msg = f"P10:0\n"
serial_port.write(msg.encode('utf-8'))
```

```python
msg = f"P10:0\n"
serial_port.write(msg.encode('utf-8'))
```
